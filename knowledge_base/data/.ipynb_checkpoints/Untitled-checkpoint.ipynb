```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
data1 = np.loadtxt('./different_icps')
```

```python
def plotfig(data1):
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(data1, cmap=plt.cm.Blues, alpha=0.8)
    fig.colorbar(heatmap)
    fig = plt.gcf()
    fig.set_size_inches(12, 12)
    ax.set_frame_on(False)
    ax.set_yticks(np.arange(data1.shape[0]) + 0.5, minor=False)
    ax.set_xticks(np.arange(data1.shape[1]) + 0.5, minor=False)
    ax.set_xlabel('Object A')
    ax.set_ylabel('Object A')
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    labels = []
    for category in range(16):
        labels.append(category)
    ax.set_xticklabels(labels, minor=False)
    ax.set_yticklabels(labels, minor=False)
    plt.xticks(rotation=90)
    ax.grid(False)
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')
    for t in ax.xaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
    for t in ax.yaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
#     fig.savefig('/home/alan/Dropbox/ComparisonA&C.pdf')
# # fig.colorbar()
```

```python
data2 = data1.copy()
num = 1e-5
data2[data2 >= num] = 1
data2[data2 < num] = 0
plotfig(data2)
```
