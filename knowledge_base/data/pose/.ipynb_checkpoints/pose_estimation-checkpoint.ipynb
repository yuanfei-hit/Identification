```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.loadtxt('../xs.txt')
y = np.loadtxt('../ys.txt')
z = np.loadtxt('../zs.txt')
pitch = np.loadtxt('../pitchs.txt')
roll = np.loadtxt('../rolls.txt')
yaw = np.loadtxt('../yaws.txt')
```

```python
num = 150
f = plt.figure(figsize=(12,8))
ax = f.add_subplot(231)
ax.hist(x, num);
plt.title('x axis')
ax = f.add_subplot(232)
ax.hist(y, num);
plt.title('y axis')
ax = f.add_subplot(233)
ax.hist(z, num);
plt.title('z axis')
ax = f.add_subplot(234)
ax.hist(pitch, num);
plt.title('pitch')
ax = f.add_subplot(235)
ax.hist(roll, num);
plt.title('roll')
ax = f.add_subplot(236)
ax.hist(yaw, num);
plt.title('yaw')
f.savefig('/home/alan/MSC_DISSERTATION/pose.pdf')
```
