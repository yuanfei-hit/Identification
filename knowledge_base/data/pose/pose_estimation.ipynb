```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.loadtxt('1/xs.txt')
y = np.loadtxt('1/ys.txt')
z = np.loadtxt('1/zs.txt')
pitch = np.loadtxt('1/pitchs.txt')
roll = np.loadtxt('1/rolls.txt')
yaw = np.loadtxt('1/yaws.txt')
```

```python
num = 20
f = plt.figure(figsize=(12,6))
ax = f.add_subplot(231)
ax.hist(x, num);
plt.title('x axis')
#plt.xlim([0.03,0.05])
ax = f.add_subplot(232)
ax.hist(y, num);
plt.title('y axis')
ax = f.add_subplot(233)
ax.hist(z, num);
plt.title('z axis')
ax = f.add_subplot(234)
ax.hist(pitch, num);
plt.title('pitch')
ax = f.add_subplot(235)
ax.hist(roll, num);
plt.title('roll')
ax = f.add_subplot(236)
ax.hist(yaw, num);
plt.title('yaw')
plt.show()
#f.savefig('/home/alan/Dropbox/figures/pose11.pdf')
```
