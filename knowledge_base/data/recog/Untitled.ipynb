```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
data = np.loadtxt('whole_slide_16.pcd')
#data = np.loadtxt('I.pcd')
#print data
data = data[:,0:2]
#print data
plt.plot(data[:,0], data[:,1],'.')
```

```python
def kernel(x, a, b):
    num = x.shape[0]
    res = np.ones((num, num))
    for i in range(num):
        for j in range(num):
            res[i,j] = a*np.exp(-1/b**2*(x[i,0] - x[i,1])**2)
    return res
```

```python
res = kernel(data, 1, 1)
x1 = np.arange(res.shape[0])
x2 = np.arange(res.shape[0])
x1,x2 = np.meshgrid(x1, x2)
# plt.imshow(res, cmap='hot')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(x1,x2,res)
```
