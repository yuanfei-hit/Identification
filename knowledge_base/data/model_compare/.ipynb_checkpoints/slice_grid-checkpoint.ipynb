```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
def plotfigbox(data1, name1, name2):
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(data1, cmap=plt.cm.hot, alpha=0.8)
    fig.colorbar(heatmap)
    fig = plt.gcf()
    fig.set_size_inches(6, 6)
    ax.set_frame_on(False)
    ax.set_yticks(np.arange(data1.shape[0]) + 0.5, minor=False)
    ax.set_xticks(np.arange(data1.shape[1]) + 0.5, minor=False)
    ax.set_xlabel('Object '+name2)
    ax.set_ylabel('Object '+name1)
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    labels = []
    for category in range(30):
        labels.append(category)
    ax.set_xticklabels(labels, minor=False)
    ax.set_yticklabels(labels, minor=False)
    plt.xticks(rotation=90)
    ax.grid(False)
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')
    for t in ax.xaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
    for t in ax.yaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
#     fig.savefig('/home/alan/Dropbox/ComparisonA&C.pdf')
# # fig.colorbar()
```

```python
A1 = np.loadtxt('similar_model_slice_scores/slice_score_between_similar_models_and_model_id_2.txt')
A2 = np.loadtxt('similar_model_slice_scores/slice_score_between_similar_models_and_model_id_4.txt')
```

```python
dc  = A1.copy()
score_upper = 3e-5
score_lower = 1e-5
for i in range(0, dc.shape[0]):  
    for j in range(0, dc.shape[1]):
        if (dc[i][j] >= score_upper):     
            dc[i][j] = 1
        elif (dc[i][j] <= score_lower):
            dc[i][j] = 0
        else:
            dc[i][j] = 0.5

plotfigbox(A1, 'Slice ID', 'Similar model IDs')
plotfigbox(dc, 'Slice ID', 'Similar model IDs')
print dc
```
