```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
def plotfig(data1, name1, name2):
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(data1, cmap=plt.cm.hot, alpha=0.8)
    fig.colorbar(heatmap)
    fig = plt.gcf()
    fig.set_size_inches(6, 6)
    ax.set_frame_on(False)
    ax.set_yticks(np.arange(data1.shape[0]) + 0.5, minor=False)
    ax.set_xticks(np.arange(data1.shape[1]) + 0.5, minor=False)
    ax.set_xlabel('Raw point cloud index of object '+name2)
    ax.set_ylabel('Raw point cloud index of object '+name1)
    ax.invert_yaxis()
    ax.xaxis.tick_top()
    labels = []
    for category in range(16):
        labels.append(category)
    ax.set_xticklabels(labels, minor=False)
    ax.set_yticklabels(labels, minor=False)
    plt.xticks(rotation=90)
    ax.grid(False)
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')
    for t in ax.xaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
    for t in ax.yaxis.get_major_ticks():
        t.tick1On = False
        t.tick2On = False
#     fig.savefig('/home/alan/Dropbox/ComparisonA&C.pdf')
# # fig.colorbar()
```
